# using beautifulsoup first csv

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import time
from time import sleep
from random import randint
```

```python
url='https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'
IMBD = requests.get(url)
soup=BeautifulSoup(IMBD.content)
```

```python
sl_no=[]
movie_name=[]
time_duration=[]
year=[]
ratings=[]
metascore=[]
director=[]
time.sleep(3)
```

```python
pages = np.arange(1,1750,50)
```

```python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_prv")
    soup = BeautifulSoup(page.text)
    movie_data = soup.find_all('div', attrs= {'class':"lister-item mode-advanced"})
    sleep(randint(2,6))
    for i in movie_data:
        number = i.h3.find('span', class_ ='lister-item-index unbold text-primary').text.replace('.','')
        sl_no.append(number)
    
        name = i.h3.a.text
        movie_name.append(name)
    
        time = i.p.find('span', class_ ='runtime').text.replace(' min', '')
        time_duration.append(time)
    
        release = i.h3.find('span', class_ ='lister-item-year text-muted unbold').text.replace('(','').replace(')','')
        year.append(release)

        rate = i.find('div', class_ ='inline-block ratings-imdb-rating').text.replace('\n', '')
        ratings.append(rate)

        meta = i.find('span', class_ ='metascore').text.replace(' ', '') if i.find('span', class_ ='metascore') else '--'
        metascore.append(meta)
        
    for i in soup.find_all('p', class_=""):
        director.append(i.text.replace('Director','').replace('\n', '').replace(':', '').replace('"','').replace('"','').replace(',','/').replace('    ','').split('|')[0])
        
    


```

```python
print(len(sl_no),len(movie_name),len(time_duration),len(year),len(ratings),len(metascore),len(director))
```

```python
df=pd.DataFrame({'Sl No':sl_no,'Movie Name':movie_name,'Time Duration':time_duration,'Year':year,'Ratings':ratings,'Metascore':metascore,'Director':director})
df
```

```python
df[['Director1', 'Director2', 'Director3','Director4','Director5','Director6']] = df['Director'].str.split('/',expand = True)
df
```

```python
df.to_csv('imbd.csv',index=False)
```
