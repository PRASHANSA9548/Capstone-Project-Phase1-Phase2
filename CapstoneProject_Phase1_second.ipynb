```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import time
from time import sleep
from random import randint
```

```python
url='https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'
IMBD = requests.get(url)
soup=BeautifulSoup(IMBD.content)

```

```python
movie_name=[]
votes=[]
genre=[]
gross_collection=[]
certificate=[]
popularity=[]
stars=[]
runtime=[]
```

```python
pages = np.arange(1,1750,50)
```

```python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_prv")
    soup = BeautifulSoup(page.text)
    movie_data = soup.find_all('div', attrs= {'class':"lister-item mode-advanced"})
    sleep(randint(2,6))
    for i in movie_data:
        name = i.h3.a.text
        movie_name.append(name)
    
        value = i.find_all('span', attrs = {'name': 'nv'})
    
        vote = value[0].text.replace(',','')
        votes.append(vote)
    
        gross = value[1].text if len(value) >1 else 'NA'
        gross_collection.append(gross)
    
        cast = i.p.find('span', class_ ="genre").text.replace('\n','').replace(',','/').replace(     '          ','')
        genre.append(cast)
        
        try:
            certificate.append(i.find('span',{"class":"certificate"}).text.replace(''','').replace(''','').replace('(','').replace(')',''))
        except:
            certificate.append(0)
      
        
        time = i.p.find('span', class_="runtime").text
        runtime.append(time)
        
        popular = i.find('div', class_ ='inline-block ratings-imdb-rating').text.replace('\n', '')
        popularity.append(popular)
    
    for i in soup.find_all('p', class_=""):
        stars.append(i.text.replace('Stars','').replace('\n', '').replace(':', '').replace('     ','').replace(',','/').split('|')[1])




```

```python
print(len(movie_name),len(votes),len(genre),len(gross_collection),len(certificate),len(stars),len(popularity))
```

```python
df=pd.DataFrame({'Movie_Name':movie_name,'Votes':votes,'Genre':genre,'Gross_Collection':gross_collection,'Certificate':certificate,'Stars':stars,'Popularity':popularity})
df
```

```python
df[['Genre1', 'Genre2', 'Genre3']] = df['Genre'].str.split('/',expand = True)
df[['Star1', 'Star2', 'Star3', 'Star4']] = df['Stars'].str.split('/',expand = True)
df
```

```python
df.to_csv('imbd2.csv',index=False)
```
