```python
# import liabrary
import sqlite3
```

```python
# this is connect command for giving file name 
dbcon=sqlite3.connect('TableimbdFirst_Database.db')
```

```python
cur=dbcon.cursor()
```

```python
# creating table
cur.execute("create table imdbtablefirst(Slno INT PRIMARY KEY,moviename TEXT,directorname TEXT,duration INT,ratings REAL)")
```

```python
with open('Table_imbd.csv','r') as file:
    total_records=0
    for row in file:
        cur.execute("insert into imdbtablefirst values(?,?,?,?,?)",row.split(","))
        dbcon.commit()
        total_records+=1
print(total_records,'Records Inserted')
```

```python
results=cur.execute("select * from imdbtablefirst")

for i in results:
    print(i)
```

```python
#Start Querying From First Table
# Query No 1
result=cur.execute("select * from imdbtablefirst where directorname in ('Christopher Nolan','Matt Reeves')")

for j in result:
    print(j)
```

```python
# Query No 2
result=cur.execute("select * from imdbtablefirst where duration>=140 and duration<=190")

for j in result:
    print(j)
```

```python
# Query No 3
result=cur.execute("select * from imdbtablefirst where ratings>7 order by ratings asc")

for j in result:
    print(j)
```

```python
# Query No 4
result=cur.execute("select * from imdbtablefirst order by moviename desc")

for j in result:
    print(j)
```

```python
# Query No 5
result=cur.execute("select * from imdbtablefirst where moviename like 'P%' and ratings>7")

for j in result:
    print(j)
```

```python
# this is connect command for giving second file name 
dbcon=sqlite3.connect('Second_Tableimbd_Database.db')
```

```python
# creating second table
cur.execute("create table imdbtablesecond(moviename TEXT,stars TEXT,votes INT,genre TEXT,grosscollection REAL,popularity REAL,certificate TEXT)")
```

```python
with open('Second_Table_imbd.csv',encoding='utf-8') as file:
    total_records=0
    for row in file:
        cur.execute("insert into imdbtablesecond values(?,?,?,?,?,?,?)",row.split(","))
        dbcon.commit()
        total_records+=1
print(total_records,'Records Inserted')
```

```python
results=cur.execute("select * from imdbtablesecond")

for row in results:
    print(row)
```

```python
# Query No 1
result =cur.execute("select * from imdbtablesecond where stars = 'Arnold Schwarzenegger'")

for j in result:
    print(j)
```

```python
# Query No 2
result =cur.execute("select * from imdbtablesecond order by votes desc")

for j in result:
    print(j)
```

```python
# Query No 3
result =cur.execute("select moviename,grosscollection from imdbtablesecond order by grosscollection desc")

for j in result:
    print(j)
```

```python
# Quer No 4
result =cur.execute("select moviename,grosscollection from imdbtablesecond where stars='Arnold Schwarzenegger'")

for j in result:
    print(j)
```

```python
# Query No 5
result =cur.execute("select * from imdbtablesecond where genre in ('Action/ Comedy  ')") 

for j in result:
    print(j)
```

# Subquery

```python
sql = "SELECT imdbtablefirst.slno, imdbtablefirst.moviename, imdbtablefirst.directorname, imdbtablefirst.duration, imdbtablefirst.ratings, imdbtablesecond.stars, imdbtablesecond.votes, imdbtablesecond.genre, imdbtablesecond.grosscollection, imdbtablesecond.popularity, imdbtablesecond.certificate FROM imdbtablefirst INNER JOIN imdbtablesecond ON imdbtablefirst.moviename = imdbtablesecond.moviename"
result =cur.execute(sql)
for i in result:
    print(i)
```

```python
sql = "SELECT imdbtablefirst.moviename, imdbtablefirst.directorname, imdbtablefirst.ratings, imdbtablesecond.grosscollection from imdbtablefirst left join imdbtablesecond on imdbtablefirst.moviename= imdbtablesecond.moviename where genre= 'Action  '"
result =cur.execute(sql)
for i in result:
    print(i)
```
